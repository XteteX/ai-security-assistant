```python
import pandas as pd

# Загрузить датасет
df = pd.read_csv(r"C:\Users\Марлен\Documents\ai-security-assistant\data\raw\security_logs.csv", encoding='utf-8')

# Удалить дубликаты (если есть заголовок "id,input_type,content,label" в середине)
df = df[df['id'].notna()].copy()
df = df[~df['id'].astype(str).str.contains('id', case=False)].copy()

# Пересчитать ID
df['id'] = range(1, len(df) + 1)

# Проверить и удалить пустые строки
df = df.dropna(subset=['content', 'label'])

# Сохранить очищенный датасет
output_path = r"C:\Users\Марлен\Documents\ai-security-assistant\data\raw\security_logs.csv"
df.to_csv(output_path, index=False, encoding='utf-8')

print(f"✓ Датасет очищен!")
print(f"Всего записей: {len(df)}")
print(f"Phishing: {len(df[df['label'] == 'phish'])}")
print(f"Safe: {len(df[df['label'] == 'safe'])}")
print(f"\nПример:")
print(df.head(10))
```
